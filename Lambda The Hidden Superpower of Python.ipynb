# Lambda The Hidden Superpower of Python

When we create a function in python it has to be named 

for example:

```python 
def named_func():

```

But does functions always has names ? the answer is no, therer are functions that has no name at all. They are called anonymous functions or **Lmabda nctions** 


Suppose we want to make a function that computes 3x+1

````python
def f(x):
    return 3*x+1

f(2)
````

Now let us create a lambda expression 

the syntax to create labmda expression :
**<font color="red">lambda</font> Input : <font color="cyan">Expression</font>**

Note: lambda is a keyword that says what follows is anonymous 

````python
lambda x: 3*x+1
````

As stated lambda is a keyword, it doesn't store values

To be able to return output all is needed is to assign it to a variable and call it

````python
var=lambda x: 3*x+1
var(2)
````

Lambda can take no input as well 

````python
example=lambda : "Hello, world!"
example()
````

### Lambda with multiple inputs

### Example and common use of lambda 

Suppose we have a list of authors and we want to sort this list based on the last name of each author, but note that some authers have middle name, while others have initials

```python

scifi_authors=["Issac Asimov" , " Ahmed Khaled Tawfik", "Arthuer C. Clarke" , "Frank Herbert", "H. G. Wells"]
```

the stratedy is to create lambda function that extracts last name and uses that as the sorting value in the built-in sort() function

````python
scifi_authors=["Frank Herbert","Orson Scott Card" , " Ahmed Khaled Tawfik", "Arthuer C. Clarke" ,  "H. G. Wells"]
help(scifi_authors.sort)
````

````python
# we will use split to split the full name and access the last name in lower case
# Note: 
# sorted() returns a new sorted list, leaving the original list unaffected.
# list.sort() sorts the list in-place, mutating the list indice
scifi_authors.sort(key=lambda name: name.split(" ")[-1].lower())
scifi_authors
````

### Writing a function that makes functions

suppose we are working with quaderatic functions:  

$ f(x)= a x^2 + bx + c$


````python
# Using normal functions 

def quadratic_func(x,a,b,c):
    """ Returns the function f(x)= a x^2 + bx + c"""
    return a*x**2+b*x+c

quadratic_func(0,2,3,-5)
````

````python
# Using Lambda functions 

def quadratic_func(a,b,c):
    """ Returns the function f(x)= a x^2 + bx + c"""
    return lambda x: a*x**2+b*x+c

func=quadratic_func(2,3,-5)

func(0)

# we can also create it without a variable

quadratic_func(2,3,-5)(0) # second bracket is x in lambda
````
